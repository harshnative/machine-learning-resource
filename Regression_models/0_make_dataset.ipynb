```python
from sklearn.datasets import make_regression
from IPython.display import display
import pandas
```

```python
# making a dataset of 50000 rows , 150 cols
myDataset = make_regression(n_samples=50000 , n_features=150 , n_informative=20)

# n_informative = The number of informative features, i.e., the number of features used to build the linear model used to generate the output.
```

```python
display(myDataset)
```

```python
X , Y = myDataset

# convert X into myDataFrame
myDataFrame = pandas.DataFrame(X , columns=range(len(X[0])))
```

```python
display(myDataFrame)
```

```python
# add Y to myDataFrame
myDataFrame["Y"] = Y
```

```python
display(myDataFrame)
```

```python
# get a subSample from the myDataFrame for 200 rows
subSample = myDataFrame.sample(n=1000)
display(subSample)
```

```python
myDataFrame.drop(list(subSample.index) , inplace=True)

display(myDataFrame)
```

```python
myDataFrame.to_csv("regression_dataset.csv")
```
