```python
import numpy
import pandas
from IPython.display import display
from IPython.display import clear_output
import random
import math
```

```python
# number of rows
n_samples = 10000

# number of cols
n_cols = 150

# number of cols having high weigth to the output of y
n_informative = 20
```

```python
if(n_informative > n_cols):
    raise ValueError("n_informative cannot be greator than n_cols")
```

```python
# class node for the equation 
# a(X ** b)
class Node:
    
    def __init__(self , a , b , x = None):
        self.a = a
        self.x = x
        self.b = b
```

```python
# generate n_informative b's with high powers

highPower_range = 25

n_informativeBs_list = list(numpy.random.normal(highPower_range , 2 , size=n_informative))

print(n_informativeBs_list)
```

```python
# generate n_cols - n_informative b's with low powers

lowPower_range = 10

restBs_list = list(numpy.random.normal(lowPower_range , 2 , size=n_cols - n_informative))

print(restBs_list)
```

```python
# combine and shuffle the bs list
n_informativeBs_list.extend(restBs_list)
Bs_list = sorted(n_informativeBs_list , reverse=True)
```

```python
display(Bs_list[:10])
```

```python
# generate n_cols a's
a_range = 1000

a_list = numpy.random.normal(1000 , 50 , size=n_cols)
```

```python
display(a_list[:10])
```

```python
# add a and b value to a(X ** b)

equation_list = []

for i in range(n_cols):
    newNode = Node(a_list[i] , Bs_list[i])
    equation_list.append(newNode)
```

```python
# export equation list for further use
import pickle

with open('equationFile', 'wb') as equationDump:
    pickle.dump(equation_list , equationDump)
```

```python
# print equation
def printEquation(nodeList):
    
    result = ""
        
    for i,j in enumerate(nodeList):
        
        result = result + "({} * {}) ^ {} + ".format(j.a , "X" + str(i+1) , j.b)
        
    print(result[:-2])
```

```python
printEquation(equation_list)
```

```python
# function to solve the equation
# gets the equation list and evalutes y by submission of a*(x**b)
def equationSolver(equation_list):
    result = 0
    for i in equation_list:
        result = result + (i.a * math.pow(i.x , i.b))
        
    return result
```

```python
# generate dataset
dataset = []
yValues = []

scaleMultilier = 10

import copy

for i in range(n_samples):
    clear_output(wait = True)
    print("on {} / {}".format(i , n_samples))
    tempEquationList = copy.deepcopy(equation_list)
    
    tempList = []
    
    # assign a random x value in equation list
    for node in tempEquationList:
        randNumber = random.random() * scaleMultilier
        node.x = randNumber
        tempList.append(randNumber)
    
    y = equationSolver(tempEquationList)
    yValues.append(y)
    dataset.append(tempList)
```

```python
# function to find max value in 2d array
def maxFinder(listPass):
    maxValue = listPass[0][0]
    for i in listPass:
        for j in i:
            if(j > maxValue):
                maxValue = j
    return maxValue
```

```python
# function to find min value in 2d array
def minFinder(listPass):
    minValue = listPass[0][0]
    for i in listPass:
        for j in i:
            if(j < minValue):
                minValue = j
    return minValue
```

```python
# to scale the values down in 2d array as whole using min max scaler
def minMaxScaler_cols(dataset , newMin , newMax):
    minValue = minFinder(dataset)
    maxValue = maxFinder(dataset)
    
    for i in range(len(dataset)):
        for j in range(len(dataset[i])):
            dataset[i][j] = (((dataset[i][j] - minValue) / (maxValue - minValue)) * (newMax - newMin)) + newMin

    return dataset
    
# to scale the values down in 1d array using min max scaler
def minMaxScaler(listPass , newMin , newMax):
    minValue = min(listPass)
    maxValue = max(listPass)
    
    for i in range(len(dataset)):
        listPass[i] = (((listPass[i] - minValue) / (maxValue - minValue)) * (newMax - newMin)) + newMin

    return listPass
    
```

```python
newRange = (0 , 10)

# scale down the values
dataset = minMaxScaler_cols(dataset , newRange[0] , newRange[1])
yValues = minMaxScaler(yValues , newRange[0] , newRange[1])
```

```python
# export to dataframe
colsList = list(range(n_cols))

myDataFrame = pandas.DataFrame(dataset , columns=colsList)
myDataFrame["Y"] = yValues

display(myDataFrame)
```

```python
# get a subSample from the myDataFrame for 200 rows
subSample = myDataFrame.sample(n=200)
display(subSample)
```

```python
# drop subsample from dataframe
myDataFrame.drop(list(subSample.index) , inplace=True)

display(myDataFrame)
```
